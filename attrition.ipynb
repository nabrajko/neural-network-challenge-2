## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head(10)
```

```python
y_df['Department'].value_counts()
```

```python
y_df['Attrition'].value_counts()
```

```python
# Create a list of at least 10 column names to use as X data
X_df = attrition_df.drop(columns=['Attrition', 'Department'])

# Create X_df using your selected columns

X_df.head()

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X=X_df
y_attrition = y_df['Attrition']
y_department = y_df['Department']
X_train, X_test, y_attrition_train, y_attrition_test, y_department_train, y_department_test = train_test_split(X, y_attrition, y_department)
```

```python
# Convert your X data to numeric data types however you see fit Add new code cells as necessary

X_train = pd.get_dummies(X_train, dtype=int)
X_test = pd.get_dummies(X_test, dtype=int)
```

```python
X_train.head()
```

```python
# Create a StandardScaler
X_scaler=StandardScaler()
# Fit the StandardScaler to the training data
X_scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
y_department_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder for the Department column
department_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
department_encoder.fit(np.array(y_department_train).reshape(-1, 1))


# Create two new variables by applying the encoder to the training and testing data
y_dept_train = department_encoder.transform(np.array(y_department_train).reshape(-1, 1))
y_dept_test = department_encoder.transform(np.array(y_department_test).reshape(-1, 1))
```

```python
# Create a OneHotEncoder for the Attrition column
attrition_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit the encoder to the training data
attrition_encoder.fit(np.array(y_attrition_train).reshape(-1, 1))

# Create two new variables by applying the encoder to the training and testing data
y_attr_train=attrition_encoder.transform(np.array(y_attrition_train).reshape(-1, 1))
y_attr_test=attrition_encoder.transform(np.array(y_attrition_test).reshape(-1, 1))
```

## Part 2: Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data.
col_num=X_train.shape[1]

# Create the input layer
input_layer = layers.Input(shape=(col_num,), name='input_features')


# Create at least two shared layers
shared_layer_1 = layers.Dense(units=32, activation='relu', name='shared_layer_1')(input_layer)
shared_layer_2 = layers.Dense(units=16, activation='relu', name='shared_layer_2')(shared_layer_1)





```

```python
# Create a branch for Department with a hidden layer and an output layer

dept_branch_hidden=layers.Dense(units=16, activation='relu', name='dept_branch_hidden')(shared_layer_2)


# Create the output layer

dept_branch_output=layers.Dense(units=y_dept_train.shape[1], activation='softmax', name='dept_branch_output')(dept_branch_hidden)

```

```python
# Create a branch for Attrition with a hidden layer and an output layer
attr_branch_hidden=layers.Dense(units=16, activation='relu', name='attr_branch_hidden')(shared_layer_2)


# Create the output layer

attr_branch_output=layers.Dense(units=y_attr_train.shape[1], activation='sigmoid', name='attr_branch_output')(attr_branch_hidden)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[attr_branch_output, dept_branch_output])
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss={'dept_branch_output': 'categorical_crossentropy', 'attr_branch_output': 'binary_crossentropy'},
              metrics={'dept_branch_output': 'accuracy', 'attr_branch_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
model_training=model.fit(
    X_train_scaled,
    {'attr_branch_output': y_attr_train, 'dept_branch_output': y_dept_train},
    epochs=100,
    batch_size=35,
    validation_split=0.2
)
```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(X_test_scaled, {'attr_branch_output': y_attr_test, 'dept_branch_output': y_dept_test})
test_results
```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy: {test_results[4]*100:.2f}%")
print(f"Attrition Accuracy: {test_results[3]*100:.2f}%")
```
